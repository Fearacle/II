Matplotlib поставляется с настройками по умолчанию, которые позволяют настраивать различные свойства. Вы можете управлять значениями по умолчанию почти каждого свойства в matplotlib: размер фигуры, количество точек на дюйм, ширина линии, цвет и стиль осей, свойства оси и сетки, свойств текста и шрифта и так далее. Значения по умолчанию для matplotlib достаточно хороши в большинстве случаев и изменять некоторые свойства нужно только для конкретных случаев.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X,C)
plt.plot(X,S)

plt.show()
```

В приведенной ниже ячейке реализованы все параметры фигуры, которые влияют на внешний вид графика. Настройки были явно установлены на значения по умолчанию, но теперь можно в интерактивном режиме играть со значениями, чтобы изучить их влияние.

```python
# импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

# Создать новую фигуру размером 8x6 точек, используя 100 точек на дюйм
plt.figure(figsize=(8,6), dpi=100)

# Создание нового заголовка
plt.subplot(111)

X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)

# Построение Косинуса синим цветом с непрерывной линией шириной 1 (пиксели)
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")

# Синус участка с использованием зеленого цвета с непрерывной линией шириной 1 (пикселей)
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")

# Установить лимит для X
plt.xlim(-4.0,4.0)

# Установить количество тиков для X
plt.xticks(np.linspace(-4,4,9,endpoint=True))

#Установить лимит для Y
plt.ylim(-1.0,1.0)

# Установить количество тиков для Y
plt.yticks(np.linspace(-1,1,5,endpoint=True))

# Сохранить рисунок с разрешением 72 dpi
# savefig("figure.png",dpi=72)

# Отобразить результат
plt.show()
```

Добавим в верхнем левом углу легенду. Для этого требуется только добавления ключевой метки аргумента (которая будет использоваться в поле легенды) к командам графика.

```python
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="Косинус")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label="Синус")

plt.legend(loc='upper left', frameon=False)

# Отобразить результат
plt.show()
```

Таким образом, **Matplotlib** позволяет гибко настраивать отображение графиков.

Еще один бесплатный пакет, основанный на этой библиотеке визуализации данных - это **Seaborn**, которая предоставляет высокоуровневый интерфейс для рисования статистической графики.

Рассмотрим простой пример:

Когда вы работаете с Seaborn, вы можете либо использовать один из встроенных наборов данных, которые имеются в библиотеке, либо можно загрузить Pandas DataFrame. Чтобы начать работать со встроенным набором данных Seaborn, вы можете использовать функцию load_dataset (), как показано в ячейке нижне. Чтобы получить обзор имеющихся наборов данных перейдите по ссылке https://github.com/mwaskom/seaborn-data.

```python
# Импортируем необходимые библиотеки
import matplotlib.pyplot as plt
import seaborn as sns

# Загружаем данные
tips = sns.load_dataset("tips")

# Создаем необходимый формат
sns.violinplot(x = "total_bill", data=tips)

# Отображаем
plt.show()
```

В большинстве случаев преимущество визуализации данных заключается в том, что работа будет происходить с собственными данными, а не со встроенными наборами данных библиотеки Seaborn. Seaborn лучше всего работает с Pandas DataFrames и массивами, которые содержат целый набор данных. 

**Помните**, что DataFrames - это способ хранения данных в прямоугольных сетках, которые легко просматривать. Каждая строка этих сеток соответствует измерениям или значениям экземпляра, в то время как каждый столбец представляет собой вектор, содержащий данные для определенной переменной. Это означает, что строки DataFrame не обязательно должны , но могут содержать значения одного и того же типа: они могут быть числовыми, символьными, логическими и т. Д. DataFrames, как уже известно, поставляются с библиотекой Pandas и определяются как двумерные маркированные структуры данных со столбцами потенциально разных типов.

### Тестовый пример
Рассмотрим на основе набора данных из ресторана о оставленных чаевых
Имеются следующие признаки:
* Общий счет (Total bill)
* Чаевые (Tips)
* Пол (Sex)
* Курящий (Smoker)
* День недели (Day)
* Время (Time)
* Количество (Size)

```python
import pandas as pd
data=pd.read_csv('tips.csv')

# Выводим первые 5 записей набора данных
data.head()
```

```python
# Смотрим на количество имеющихся данных
len(data)
```

```python
# Реализуем график, что бы посмотреть на взаимосвязь чаевых от общего счета
sns.regplot(x='tip', y='total_bill', data=data, ci=None)
plt.show()
```

```python
# Строим гистограмму для общего счета в ресторане
plt.hist(data['total_bill'])
plt.show()
```

```python
# Изменим свойства hist
sns.distplot(data['total_bill'], kde=False, rug=True)
plt.show()

# Таким образом получим более наглядую гистограмму
```

```python
# Строим boxplot для Общего счета и дня недели
sns.set_style('darkgrid')
sns.boxplot(x='day', y='total_bill', data=data, palette='Blues')
plt.show()
```

```python
# Строим boxplot для Пола и оставленных чаевых
sns.boxplot(x='sex', y='tip', data=data, palette='Blues')
plt.show()
```

```python
# Добавляем параметры и сохраняем рисунок
fig, ax= plt.subplots()
ax.scatter(data[data['sex']=='Male']['tip'], data[data['sex']=='Male']['total_bill'], color='blue', label='Муж')
ax.scatter(data[data['sex']=='Female']['tip'], data[data['sex']=='Female']['total_bill'], color='green', label='Жен')
ax.set_title('Общий счет и чаевые с распределением по полу', fontdict={'fontsize': 14, 'fontweight': 'bold'})
ax.set_ylabel('Общий счет')
ax.set_xlabel('Чаевые')
ax.legend()
plt.show()

fig.savefig('tips.png', format='png', dpi=250)
```
